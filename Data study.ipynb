## 3.Inference and Features Nomenclature:
### A. Here we are trying to collect exactly 5 kinds of data i.e.
1. MI_dir: Refers to the mutual information between the direction of the packet and label.
2. H: Entropy of the packet's length.
3. HH: Entropy of the packet's direction.
4. HH_jit: Jitter of the packet's direction
5. HpHp: Traffic measure from the packet's host + port to the packet's destination host + port

### B. Here we have also considered different window sizes (based on time):
L: Lambda 
> refers to the decay factor used in the damped window

> refers to how much recent history of the stream is captured in these statistics

1. L5
2. L3
3. L1
4. L0.1
5. L0.01

Q: But why have all these different time windows?
> Different time windows capture different aspects of the network traffic statistics.
> 
> For instance, shorter time windows such as L5 and L3 capture the more recent history of the stream and are useful for detecting short-lived attacks such as port scans or DDoS attacks 5.
> 
> On the other hand, longer time windows such as L0.01 capture more long-term patterns in the traffic and are useful for detecting slow attacks such as brute-force attacks or botnet infections 5.

### C. Different mathematical parameters which are measured here:
1. Weight [MI, H, HH, HH_jit, HpHp]: Weight of the stream or Number of items observed in recent history.
2. Mean [MI, H, HH, HH_jit, HpHp]: Average value of the feature across all the data points
3. Variance [MI, H, HH_jit]: Variance of the feature.
4. Std [HH, HpHp]: Std refers to standard deviation.
5. Magnitude [HH, HpHp]: Root squared sum of the means of the 2 streams.
6. Radius [HH, HpHp]: 
7. Cov [HH, HpHp]: Approximated covariance between 2 streams
8. PCC [HH, HpHp]: Pearson Correlation Coefficient (similar to cov)

<!-- Why do we measure all these different mathematical parameters for a single thing? -->

Q: Difference between Cov and PCC?

Ans: Covariance and Pearson correlation coefficient (PCC) are both measures of the relationship between two variables.
Covariance measures how much two variables change together, while PCC measures the strength and direction of the linear relationship between two variables.
Covariance is a measure of how much two variables vary together, while PCC is a measure of how much two variables are related on a scale from -1 to 1.
Covariance is affected by the scale of the variables, while PCC is not affected by the scale of the variables.s 12.

```python
# So now we have a list of data frames of all the files, next step is to take samples of each of them

def type(x):
    if 'benign' in x:
        return 0
    elif 'gafgyt.combo' in x:
        return 1
    elif 'gafgyt.junk' in x:
        return 2
    elif 'gafgyt.scan' in x:
        return 3
    elif 'gafgyt.tcp' in x:
        return 4
    elif 'gafgyt.udp' in x:
        return 5
    elif 'mirai.ack' in x:
        return 6
    elif 'mirai.scan' in x:
        return 7
    elif 'mirai.syn' in x:
        return 8
    elif 'mirai.udpplain' in x:
        return 9
    else :
        return 10

def danger(x):
    if 'benign' in x:
        return 0
    else:
        return 1

dfs = []
for x in list_of_files:
    if x in ['README.md', 'features.csv', 'device_info.csv', 'data_summary.csv']:
        continue
    new_df = pd.read_csv('../data/' + x)
    new_df['file'] = x
    new_df['type'] = type(x)
    new_df['danger'] = danger(x)
    dfs.append(new_df)
```

```python
for i in range (len(dfs)):
    dfs[i] = dfs[i].sample(n=300)
```

```python
allSamples = pd.concat(dfs, axis=0, ignore_index=True)
allSamples.sample(100)
```
